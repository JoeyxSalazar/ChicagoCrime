## This is going to explore the domestic violence cases in the City of Chicago

```python
import pandas as pd
df = pd.read_csv('data/ChicagoArrests.csv')
```

#### Lets filter to cases where there is domestic violence involved

```python
dom_abusers = df[df["CHARGES STATUTE"].str.contains("720 ILCS 5.0/12-3.2-A-1|720 ILCS 5.0/12-3.3-A-1|720 ILCS 5.0/12-3.4-A-1|720 ILCS 5.0/12-3.5-A-1|720 ILCS 5.0/12-3.8-A-1|720 ILCS 5.0/12-3.9-A-1", case=False, na=False)]
#Add new columns to expose perpetrator
dom_abusers.insert(2, "NAME", None)
dom_abusers.insert(3, "AGE", None)
dom_abusers.insert(4, "ARREST LOCATION", None)
dom_abusers.head()
```

# Percent of crime from 2014 - 2025 that is Domestic violence

```python
p = (dom_abusers.shape[0] / df.shape[0]) * 100
print(p)
```
